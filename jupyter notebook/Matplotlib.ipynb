```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.array([1, 2, 3, 4, 5])

# Creating a figure and axis
fig, ax = plt.subplots(figsize=(10, 4))

# Plotting the lines
ax.plot(x, x**2, label='x^2', lw=2)
ax.plot(x, x**3, label='x^3', lw=2)

# Setting x-ticks and their labels
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)

# Adding a legend
ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Creating a figure and axis
fig, ax = plt.subplots()

# Plotting the data
ax.plot(x, y)

# Setting custom xticks and xtick labels
xticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
xtick_labels = [r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']
ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels, fontsize=12)

# Setting custom yticks and ytick labels
yticks = [-1, -0.5, 0, 0.5, 1]
ytick_labels = ['-1', '-0.5', '0', '0.5', '1']
ax.set_yticks(yticks)
ax.set_yticklabels(ytick_labels, fontsize=12)

# Adding labels and title
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_title('Sine Wave')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
xx = np.linspace(0, 10, 50)
n = np.arange(1, 10)

# Creating subplots
fig, axes = plt.subplots(1, 4, figsize=(12, 3))

# Scatter plot
axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")

# Step plot
axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")

# Bar plot
axes[2].bar(n, n**2)
axes[2].set_title("bar")

# Fill between plot
axes[3].fill_between(n, n**2, n**3, color='lightblue', alpha=0.6)
axes[3].set_title("fill_between")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

# Example data
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
y = np.linspace(-2 * np.pi, 2 * np.pi, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.cos(Y)

# Creating a figure and axis
fig, ax = plt.subplots()

# Creating a pseudocolor plot
p = ax.pcolor(X / (2 * np.pi), Y / (2 * np.pi), Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())

# Adding a colorbar
cb = fig.colorbar(p, ax=ax)

plt.show()
```
